# Advance operation with Business examples

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Data structure : [restaurant_id,2021,2022,2023,2024]
sales_data = np.array([
    [1, 5000, 7000, 8000, 4000],      #paradise biryani
    [2, 6000, 6500, 7000, 7500],      #beiging bites
    [3, 4000, 4500, 5000, 5500],        #spice hub
    [4, 8000, 8500, 7000, 9500],        #curry corner
    [5, 3000, 3500, 4000, 4500]         #tandoori treats
])

print("===Zomato Sales Data Analysis===")
print("\n sales data shape:", sales_data.shape)  #shape of the data
print("\n sample data for 1st 3 restaurant:\n", sales_data[:3])  #printing the data
print("\n print all the column except restaurant id:\n", sales_data[:,1:])  #print all the column except restaurant id
```

```python
# total sales for each year
print(np.sum(sales_data,axis=0))  #axis=0 means column wise operation
yearly_total=np.sum(sales_data[:,1:],axis=1) #print along the y axis (row wise operation)
print("\n yearly total sales from 2021 to 2024:",yearly_total)

#minimum sales per restaurant
min_sales=np.min(sales_data[:,1:],axis=1)
print("\n minimum sales per restaurant from 2021 to 2024:",min_sales) # calculating minimum sales per restaurant y axis


#maximum sales per restaurant
max_sales=np.max(sales_data[:,1:],axis=0)
print("\n maximum sales per year from 2021 to 2024:",max_sales) # calculating maximum sales per year x axis

3# average sales per restaurant
avg_sales=np.mean(sales_data[:,1:],axis=1)
print("\n average sales per restaurant from 2021 to 2024:",avg_sales)
```

# commululative sales

```python
cumsum=np.cumsum(sales_data[:,1:],axis=1)
print(cumsum)

plt.figure(figsize=(10,6))
plt.plot(np.mean(cumsum,axis=0))
plt.title("Cumulative Sales Over Years (Average Across Restaurants)")
plt.xlabel("Years (2021 to 2024)")  
plt.ylabel("Cumulative Sales")
plt.grid(True)
plt.show()
```

```python
vector1 = np.array([1,2,3,4,5])
vector2 = np.array([6,7,8,9,10])

print("Vector Addition:", vector1 + vector2)

print("Vector multiplication:", vector2 * vector1)

print("Dot Product:", np.dot(vector1, vector2))

angle= np.arccos(np.dot(vector1, vector2) / (np.linalg.norm(vector1) * np.linalg.norm(vector2)))
print("Angle between vectors (radians):", angle)
```
